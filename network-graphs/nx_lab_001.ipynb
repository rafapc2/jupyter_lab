```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_node(1)
```

```python
G.add_nodes_from([2,3,4,5,6])
```

```python
nx.draw(G)
```

```python
nx.draw(G, with_labels=True)
```

```python
G.add_edge(1,2)
```

```python
G.add_edge(1,3)
G.add_edge(1,4)
G.add_edge(1,5)
G.add_edge(1,6)
nx.draw(G, with_labels=True)
```

```python
print(f"{G.number_of_edges()=}")
print(f"{G.size()=}")


print(f"{G.number_of_nodes()=}")
print(f"{G.order()=}")
```

```python
print(f"{nx.degree(G)=}")

print(f"{nx.density(G)=}")

#red no dirigida cada nodo tiene 2 conexiones
density = G.number_of_edges()*2/(G.number_of_nodes()*(G.number_of_nodes()-1))

print(f"{density=}")

```

```python
#maxima distancia del grafo

nx.diameter(G)
```

```python
#red completa (todos con todos)
H = nx.complete_graph(6)
nx.draw(H, with_labels=True)
```

```python
print(f"{H.number_of_edges()=}")
print(f"{H.size()=}")


print(f"{H.number_of_nodes()=}")
print(f"{H.order()=}")

print()
print(f"{nx.degree(H)=}")

print(f"{nx.density(H)=}")
print()

#maxima distancia del grafo

print(f"{nx.diameter(H)=}")
```

```python
#red no compleja
I = nx.gnp_random_graph(5, 0.60)
nx.draw(I, with_labels=True)
```

```python
print(f"{I.number_of_edges()=}")
print(f"{I.size()=}")


print(f"{I.number_of_nodes()=}")
print(f"{I.order()=}")

print()
print(f"{nx.degree(I)=}")

print(f"{nx.density(I)=}")
print()

#maxima distancia del grafo
try:
    print(f"{nx.diameter(I)=}")
except:
    print("Para este grafo no se puede calcualar el diamtro, hay nodos desconectados")
```
