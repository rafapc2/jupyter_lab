```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
G = nx.barabasi_albert_graph(100, 2) #random g


```

```python
nx.draw_spring(G, with_labels=True)
```

```python
nx.draw(G, with_labels=True)
```

```python
plt.hist([v for k,v in nx.degree(G)])
```

```python
plt.hist(nx.centrality.betweenness_centrality(G).values())
```

```python
plt.hist(nx.centrality.closeness_centrality(G).values())
```

```python
print(f"{G.number_of_edges()=}")
print(f"{G.size()=}")


print(f"{G.number_of_nodes()=}")
print(f"{G.order()=}")

print()
print(f"{nx.degree(G)=}")

print(f"{nx.density(G)=}")
print()

#maxima distancia del grafo
try:
    print(f"{nx.diameter(G)=}")
except:
    print("Para este grafo no se puede calcualar el diamtro, hay nodos desconectados")
```

```python
#pandas

nodes = list(range(100))
df = pd.DataFrame({'from': np.random.choice(nodes,100),
                   'to':   np.random.choice(nodes, 100)

                })

display(df.sample(20))
```

```python
I = nx.from_pandas_edgelist(df, source='from', target='to')
```

```python
nx.draw(I, with_labels=True)
```

```python
print(f"{I.number_of_edges()=}")
print(f"{I.size()=}")


print(f"{I.number_of_nodes()=}")
print(f"{I.order()=}")

print()
print(f"{nx.degree(I)=}")

print(f"{nx.density(I)=}")
print()

#maxima distancia del grafo
try:
    print(f"{nx.diameter(I)=}")
except:
    print("Para este grafo no se puede calcualar el diametro, hay nodos desconectados")
```
